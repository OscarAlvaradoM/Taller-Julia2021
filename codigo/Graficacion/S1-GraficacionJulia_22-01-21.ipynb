#### Subplots

```julia
x = 1:10
y = rand(10, 4)
plot(x, y, layout = (4, 1), c = ["blue" "red" "green" "purple"])
```

```julia
x = 1:10
y = rand(10, 4)
p1 = plot(x, y)
p2 = scatter(x, y, shape = :cross)
p3 = plot(x, y, xlabel = "¡Esta está tuneada!", lw = 4, title = "Subtítulo", line = :dash)
p4 = histogram(x, y)
plot(p1, p2, p3, p4, layout = (2, 2), legend = false)
```

```julia
# Ahora intentemos con funciones: 
x = 1:0.1:20
f(x) = x*sin(x) - 0.2x
plot(x, f.(x), c = "pink", lw = 4, title = "Graficando con funciones", xlabel = "x", ylabel = "f(x)")
```

```julia
# Nos podemos poner ingeniosos
n = 400
t = range(0, 2π, length = n)
x = 16sin.(t).^3
y = 13cos.(t) .- 5cos.(2t) .- 2cos.(3t) .- cos.(4t)
plot([x 1.02x], [y 1.02y], c = ["red" :red], lw = 4, title = "Corazón", xlabel = "x", ylabel = "f(x)",
    background_color = :pink, legend = false, fill = (1, :red)) # Vemos que el color también puede ser un símbolo
annotate!(0, 0, "<Inserte nombre>")
```

```julia
# A ver, una en 3D
x = 1:0.1:20
y = sin.(x)
z = cos.(x)
plot(x,y,z)
plot!(2x,2y,2z)
```

```julia
# Se pueden crear figuras propias y poner como bullets! 

verts = [(-1.0, 1.0), (-1.28, 0.6), (-0.2, -1.4), (0.2, -1.4), (1.28, 0.6), (1.0, 1.0), (-1.0, 1.0), (-0.2, -0.6), (0.0, -0.2), (-0.4, 0.6), (1.28, 0.6), (0.2, -1.4), (-0.2, -1.4), (0.6, 0.2), (-0.2, 0.2), (0.0, -0.2), (0.2, 0.2), (-0.2, -0.6)]
x = 0.1:0.2:0.9
y = 0.7 * rand(5) .+ 0.15
plot(x, y, line = (3, :dash, :lightblue), marker = (Shape(verts), 30, RGBA(0, 1, 0, 0.4)), legend = false)
```

#### Animaciones

```julia
function rotar_2D(θ, M)
    R = [cos(θ) -sin(θ);
        sin(θ) cos(θ)]
    M′ = R * M
    return M′
end
```

```julia
x = [i[1] for i ∈ verts]
y = [j[2] for j ∈ verts]
vertices = [x, y]
anim = @animate for θ in 0:0.1:2π
    rot = rotar_2D(θ, vertices)
    plot(rot[1,:], rot[2,:], key = false, show = false, xlim = [-2,2], ylims = [-2,2], c = :green)
    end
gif(anim,"../imgs/anim.gif")
```
